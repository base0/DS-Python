```python
import pandas as pd

df = pd.read_csv('tips.csv')
df.head()
```

```python
df.plot()
```

```python
df.plot(subplots=True)
```

```python
df['day'].value_counts().plot(kind='bar', color=['grey','darkblue','grey','grey',])
```

```python
df['tip'].plot(kind='hist', bins=10)
```

```python
df.plot(kind='scatter', x='total_bill', y='tip', alpha=0.3)
```

```python
df.corr()
```

```python
df_m = df[df.sex == 'Male']
df_w = df[df.sex == 'Female']
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)  # row, col, index
df_m.plot(title='Man',   ax=plt.gca(), kind='scatter', x='total_bill', y='tip', alpha=0.3)
plt.subplot(1,2,2)
df_w.plot(title='Woman', ax=plt.gca(), kind='scatter', x='total_bill', y='tip', alpha=0.3)
```

```python
ax = df_m.plot(  kind='scatter', x='total_bill', y='tip', c='b', alpha=0.3)
df_w.plot(ax=ax, kind='scatter', x='total_bill', y='tip', c='r', alpha=0.3)
```
