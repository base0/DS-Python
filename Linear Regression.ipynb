# Linear Regression

# Prepare Data

```python
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
import numpy as np

X = np.arange(-4, 4, 0.25)
y = X**4 + X**3 - 13*(X**2) - X + 12
```

```python
plt.plot(X, y)
```

```python
plt.scatter(X, y)
```

# Linear Regression

Learn

```python
from sklearn import linear_model

X2D = np.reshape(X, (len(X), 1))
lm = linear_model.LinearRegression()
lm.fit(X2D, y)
```

Test the performance using MAE (mean absolute error) & RMSE (root mean square error).

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

y_pred = lm.predict(X2D)
print(mean_absolute_error(y, y_pred))
print(mean_squared_error(y, y_pred) ** 0.5)
```

```python
plt.plot(X, y, 'k')
plt.plot(X, y_pred)
```

```python
print(lm.coef_)
print(lm.intercept_)
```
